```python
from transformers import pipeline

# Load HeliSanKY model for Arabic to English translation
translator_ar_to_en = pipeline("translation", model="Helsinki-NLP/opus-mt-ar-en")

# Load AraT5 model for Arabic dialects translation
translator_arabic_dialects = pipeline("translation", model="PRAli22/arat5-arabic-dialects-translation")



```

# From Egyptian arabic to English pipeline

after adding the fintuned model we will get more efficient results

```python
def from_egyptian_to_english_pipeline(text):
    translated_from_3amya_to_arabic = translator_arabic_dialects(text, max_length=512)[0]['translation_text']

    translated_arabic_to_en = translator_ar_to_en(translated_from_3amya_to_arabic, max_length=512)[0]['translation_text']

    return translated_arabic_to_en

# Example
text = "انا روحت امبارح عندك فى البيت"
translated_text = from_egyptian_to_english_pipeline(text)
print("Translated text:", translated_text)
```

# From English to Arabic pipeline

we can fine tune arat5 on egyptian arabic to make the pipeline from english to egyptian arabic

```python
def arabic_translation_pipeline(text):
    # Translate Arabic 3maya to English using HeliSanKY model
    translated_to_en = translator_ar_to_en(text, max_length=512)[0]['translation_text']
    # Translate the English text back to Arabic using AraT5 model
    translated_back_to_arabic = translator_arabic_dialects(translated_to_en, max_length=512)[0]['translation_text']
    return translated_back_to_arabic

text = "i went to your home yesterday."
translated_text = arabic_translation_pipeline(text)
print("Translated text:", translated_text)
```
